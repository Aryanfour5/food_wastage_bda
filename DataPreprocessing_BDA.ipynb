```python
# Cell 1: Upload and Load Data
from google.colab import files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Upload CSV file
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Load raw data
df_raw = pd.read_csv(filename)
print(f"Raw data shape: {df_raw.shape}")
df_raw.head(10)
```

```python
# Cell 2: Inspect Data
print("=== DATA OVERVIEW ===")
print(df_raw.info())
print("\n=== MISSING VALUES ===")
print(df_raw.isnull().sum())
print("\n=== SAMPLE DATA ===")
df_raw.head()
```

```python
# Cell 3: Data Cleaning & Preprocessing

# Drop rows with missing critical columns
df = df_raw.dropna(subset=['commodity', 'year', 'loss_percentage'])

# Clean loss_percentage - remove % signs and convert to numeric
df['loss_percentage_clean'] = df['loss_percentage'].apply(
    lambda x: float(str(x).replace('%', '').strip()) if pd.notna(x) else np.nan
)

# Convert year to integer
df['year'] = pd.to_numeric(df['year'], errors='coerce').astype('Int64')

# Convert numeric columns
df['loss_quantity'] = pd.to_numeric(df['loss_quantity'], errors='coerce').fillna(0)
df['sample_size'] = pd.to_numeric(df['sample_size'], errors='coerce').fillna(0)
df['m49_code'] = pd.to_numeric(df['m49_code'], errors='coerce').fillna(0).astype(int)

# Fill missing categorical values
df['country'] = df['country'].fillna('Unknown')
df['region'] = df['region'].fillna('Unknown')
df['activity'] = df['activity'].fillna('Unknown')
df['food_supply_stage'] = df['food_supply_stage'].fillna('Unknown')
df['treatment'] = df['treatment'].fillna('UNK')
df['cause_of_loss'] = df['cause_of_loss'].fillna('UNK')
df['method_data_collection'] = df['method_data_collection'].fillna('Unknown')

# Remove outliers - loss percentage should be 0-100%
df = df[(df['loss_percentage_clean'] >= 0) & (df['loss_percentage_clean'] <= 100)]

# Select final columns
cleaned_df = df[[
    'm49_code', 'country', 'region', 'cpc_code', 'commodity', 'year',
    'loss_percentage_clean', 'loss_quantity', 'activity', 'food_supply_stage',
    'treatment', 'cause_of_loss', 'sample_size', 'method_data_collection'
]].rename(columns={'loss_percentage_clean': 'loss_percentage'})

print(f"✅ Cleaned data shape: {cleaned_df.shape}")
cleaned_df.head(10)
```

```python
# Cell 4: Data Quality Report

print("=== DATA QUALITY REPORT ===")
print(f"Total Records: {len(cleaned_df)}")
print(f"Unique Commodities: {cleaned_df['commodity'].nunique()}")
print(f"Unique Countries: {cleaned_df['country'].nunique()}")
print(f"Unique Regions: {cleaned_df['region'].nunique()}")
print(f"Year Range: {cleaned_df['year'].min()} - {cleaned_df['year'].max()}")
print(f"Loss % Range: {cleaned_df['loss_percentage'].min():.2f}% - {cleaned_df['loss_percentage'].max():.2f}%")
print(f"\nMissing values after cleaning:")
print(cleaned_df.isnull().sum())
```

```python
# Cell 5: Visualizations

# Loss by commodity (top 10)
plt.figure(figsize=(12, 6))
top_commodities = cleaned_df.groupby('commodity')['loss_percentage'].mean().sort_values(ascending=False).head(10)
top_commodities.plot(kind='bar', color='coral')
plt.title('Top 10 Commodities by Average Loss Percentage')
plt.xlabel('Commodity')
plt.ylabel('Average Loss %')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Loss over years
plt.figure(figsize=(12, 6))
yearly_loss = cleaned_df.groupby('year')['loss_percentage'].mean()
yearly_loss.plot(kind='line', marker='o', color='blue')
plt.title('Average Food Loss % Over Years')
plt.xlabel('Year')
plt.ylabel('Average Loss %')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Loss by activity
plt.figure(figsize=(12, 6))
activity_loss = cleaned_df.groupby('activity')['loss_percentage'].mean().sort_values(ascending=False).head(10)
activity_loss.plot(kind='barh', color='green')
plt.title('Top 10 Activities by Average Loss %')
plt.xlabel('Average Loss %')
plt.ylabel('Activity')
plt.tight_layout()
plt.show()

# Loss distribution
plt.figure(figsize=(10, 6))
plt.hist(cleaned_df['loss_percentage'], bins=50, color='purple', edgecolor='black', alpha=0.7)
plt.title('Distribution of Loss Percentage')
plt.xlabel('Loss %')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Cell 6: Export Cleaned Dataset

# Save to CSV
cleaned_df.to_csv('flw_cleaned.csv', index=False)
print("✅ Cleaned dataset saved as 'flw_cleaned.csv'")

# Download the file
files.download('flw_cleaned.csv')
```
